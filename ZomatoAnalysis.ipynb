STEP 1: IMPORTING LIBRARIES

(i)Pandas is used for data manipulation and analysis
(ii) numpy is used for numerical operations
(iii) matplotlib.pyplot and seaborn is used for data visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2- Create the data Frame

```python
dataframe = pd.read_csv("Zomato data .csv")
```

```python
dataframe
```

Convert the data type of column - rate

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)

dataframe['rate'] =dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

**Type of Resturant**

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of resturant")
```

*conclusion- majority of the resturant falls in dining category*

II. votes of each type of resturant

```python
dataframe.head()
```

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes': grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of resturant",c="red",size=20)
plt.ylabel("votes",c="red",size=20)
```

**what are the rating majority of resturants have received**

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=10)
plt.title("ratings distribution")
plt.show()
```

Conclusion - The mojority resturant received rating from 3.5 -4

**Average order spending by couples**

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

conclusion - the mojority of couples prefer resturants with an aproximate cost of 300 rs

**Which mode receives maximum rating**

```python
dataframe.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

conclusion- offline order received lower rating in comparision with offline rating

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order', aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt='d')
plt.title("Heatmap ")
plt.xlabel("Online Order")
plt.ylabel("Listed In (type)")
plt.show()
```
